### Exercise 1

Compute and plot the stationary distribution of the matrix using one of the methods in `quantecon`'s `MarkovChain` object, combined with `matplotlib`.

```python
def plot_stationary(mc, stationaryDist, numIts = None):
    plt.bar(np.arange(0,8), stationaryDist)
    if numIts == None:
        plt.title("Stationary Distribution")
    else:
        plt.title("Approximate Stationary Distribution: %i Periods" %numIts)
    plt.xlabel("State")
    plt.ylabel("Probability")
    plt.show()
```

```python
mc = qe.MarkovChain(P)
stationary = mc.stationary_distributions
stationaryDist = stationary[0]
plot_stationary(mc, stationaryDist)
```

### Exercise 2

This Markov chain is asymptotically stationary and ergodic, which means that, for any sample path $\{X_t\}$ generated according to $P$, we have

$$ \lim_{n \to \infty} \frac{1}{n} \sum_{t=1}^n \mathbb 1\{X_t = j\} = \psi(j) $$

where $\psi$ is the stationary distribution and $j$ is an integer between 0 and 7 (the set of possible states).

Use this fact to compute an approximation to the stationary distribution $\psi$.  Confirm visually that your result is similar to the stationary distribution you computed in exercise 1.

You can make use of the `simulate` method in `MarkovChain`.

```python
def simulate_approx(mc, n):
    '''Approximate stationary distribution for markov chains
    that are stationary and ergodic.'''

    # Simulate path 
    path = mc.simulate_indices(ts_length=n)
    # Count the number of periods in each state
    states, count = np.unique(path, return_counts=True)
    stationary = count / n
    
    # Plot approximate stationary distribution
    
    plot_stationary(mc, stationary, n)
    
    return stationary
```

```python
simulate_approx(mc, 10000);
```

Indeed, this distribution is very similar to the stationary distribution computed above.

### Exercise 3

Ergodicity also implies that, if we simulate a large number of paths and then look at the cross section at some $T$, where $T$ is suitably large, then the empirical distribution should be close to the stationary distribution.

Confirm this by simulation and visual inspection, as above.

(In this context, the empirical distribution of a sample is the fraction of observations that take value $j$ for each $j$ in $0, \ldots, 7$.)

```python
def simulation_ergo(mc, numPaths, T):
    
    stateT = np.zeros((numPaths,1))
    for ii in range(numPaths):
        path = mc.simulate_indices(ts_length=T)
        stateT[ii] = path[T-1]
    states, count = np.unique(stateT, return_counts=True)
    
    stationary = count / numPaths
    plot_stationary(mc, stationary)
```

```python
simulation_ergo(mc, 10000, 10000)
```

Again, this distribution is very similar to the stationary distribution computed above.
